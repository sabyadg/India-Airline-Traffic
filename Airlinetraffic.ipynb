```python
#Loading Quarter 1 data
dfQ1countrywise = pd.read_csv('Table_0_Countrywise_DGCA_Q1_Jan-Mar_2017.csv', delimiter=',')
dfQ1countrywise = dfQ1countrywise.drop(dfQ1.index[[58]])
dfQ1countrywise = dfQ1countrywise.rename(columns={'NAME OF THE COUNTRY': 'Country', 'Sl. No.':'Index', 'PASSENGERS TO INDIA':'Traffic_to_India', 'PASSENGERS FROM INDIA':'Traffic_from_India','FREIGHT TO INDIA':'Freight_to_India','FREIGHT FROM INDIA':'Freight_from_India'})


dfQ1countrysort = dfQ1countrywise.sort_values(['Traffic_to_India','Traffic_from_India'],ascending=False)
dfQ1countrysort['index value'] = np.arange(0,len(dfQ1countrywise['Index']))
dfQ1countrysortsample = dfQ1countrysort[0:25]
dfQ1countrysortsample
```

```python
#Loading Quarter 2 data
dfQ2 = pd.read_csv('Table_1_Airlinewise_DGCA_Q2_APR-JUN_2017.csv', delimiter=',')
dfQ2 = dfQ2.drop(dfQ2.index[[5,97,98]])
dfQ2 = dfQ2.rename(columns={'NAME OF THE AIRLINE': 'Airline', 'SL.No.':'Index', 'PASSENGERS TO INDIA':'Traffic_to_India', 'PASSENGERS FROM INDIA':'Traffic_from_India','FREIGHT TO INDIA':'Freight_to_India','FREIGHT FROM INDIA':'Freight_from_India'})


dfQ2sort = dfQ2.sort_values(['Traffic_to_India','Traffic_from_India'],ascending=False)
dfQ2sort['index value'] = np.arange(0,len(dfQ2['Index']))
dfQ2sortsample = dfQ2sort[0:25]
dfQ2sortsample[0:10]
```

```python
def plotairlinetraffic (dfQ2sort):
    
    font = {'family' : 'helvetica',
        'weight' : 'bold',
        'size'   : 22}

    plt.rc('font', **font)
    
    #plotting correlation between in and out traffic
    fig = plt.figure()
    fig.set_figwidth(5)
    fig.set_figheight(5)
    plt.plot(dfQ2sort['Traffic_from_India']/1000,dfQ2sort['Traffic_to_India']/1000, 'x')
    plt.xlabel('Traffic from India (in $10^3$)')
    plt.ylabel('Traffic to India (in $10^3$)')
    plt.xticks(np.arange(min( min(dfQ2sort['Traffic_from_India']/1000), 0),max(max(dfQ2sort['Traffic_from_India']/1000), 2500), 500) )
    plt.yticks(np.arange(min(min(dfQ2sort['Traffic_to_India']/1000), 0),max(max(dfQ2sort['Traffic_to_India']/1000), 2500), 500))
    plt.grid()
    plt.show()
    
    #plotting bar plot of traffic to and from India as horizontal stacked chart
    fig = plt.figure()
    fig.set_figwidth(35)
    fig.set_figheight(20)
    colors ='rgbwmc'
    p1 = plt.bar( dfQ2sort['index value'],dfQ2sort['Traffic_to_India']*0.001, width = 0.5, color = colors[0])
    p2 = plt.bar( dfQ2sort['index value'], dfQ2sort['Traffic_from_India']*.001 ,width = 0.5,
                 bottom=dfQ2sort['Traffic_to_India']*.001, color = colors[1])

    plt.ylabel('Traffic Volume (in $10^3$)')
    plt.title('Airline Traffic Volume for Q2 (Apr-June)  ')
    plt.xticks(dfQ2sort['index value'], dfQ2sort['Airline'])

    plt.legend((p1[0], p2[0]), ('Traffic to India', 'Traffic from India'))
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=90)
    plt.tick_params(labeltop=False, labelright=True)
    plt.grid()
    plt.show(p1)
    
     
    #plotting bar plot of traffic to and from India (individually)
    fig = plt.figure()
    fig.set_figwidth(20)
    fig.set_figheight(40)
    colors ='rgbwmc'
    p1 = plt.barh( dfQ2sort['index value'], dfQ2sort['Traffic_to_India']/1000 , color = colors[0])

    plt.xlabel('Traffic Volume to India (in $10^3$)')

    plt.yticks(dfQ2sort['index value'], dfQ2sort['Airline'])
    #plt.xticks((0, 1000))
    plt.tick_params(labeltop=True, labelright=False)
    locs, labels = plt.yticks()
    plt.setp(labels, rotation=0)
    plt.grid()
    plt.show(p1)

    fig = plt.figure()
    fig.set_figwidth(20)
    fig.set_figheight(40)
    colors ='rgbwmc' 
    p3 = plt.barh( dfQ2sort['index value'], dfQ2sort['Traffic_from_India']/1000 , color = colors[1])

    plt.xlabel('Traffic Volume from India (in $10^3$)')

    plt.yticks(dfQ2sort['index value'], dfQ2sort['Airline'])
    #plt.xticks((0, 1000))
    plt.tick_params(labeltop=True, labelright=False)
    locs, labels = plt.yticks()
    plt.setp(labels, rotation=0)
    plt.grid()
    plt.show(p3)
    
    dfQ2sort = dfQ2sort.sort_values(['index value'] , ascending = False)
    index = dfQ2sort['index value']
    data = np.array([dfQ2sort['Traffic_to_India']/1000, dfQ2sort['Traffic_from_India']/1000])
    y_pos = np.arange(len(index))

    fig = plt.figure(figsize=(10,40))
    ax = fig.add_subplot(111)

    colors ='rgbwmc'
    patch_handles = []
    left = np.zeros(len(index)) # left alignment of data starts at zero
    for i, d in enumerate(data):
        patch_handles.append(ax.barh(y_pos, d, 
          color=colors[i%len(colors)], align='center', 
          left=left))
        # accumulate the left-hand offsets
        left += d

    ax.set_yticks(y_pos)
    ax.tick_params(labeltop=True, labelright=False)
    ax.set_yticklabels(dfQ2sort['Airline'])

    ax.set_xlabel('Traffic Volume (in $10^3$)')
    plt.grid()
    plt.show()

    return()
```
